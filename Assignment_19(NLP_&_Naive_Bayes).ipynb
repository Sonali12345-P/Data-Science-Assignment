```python
import re
import string
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def clean_text(text):
  text = re.sub(f'[{re.escape(string.punctuation)}0-9]', '', text)

  text = text.lower()

  words = word_tokenize(text)

  stop_words = set(stopwords.words('english'))
  words = [word for word in words if word not in stop_words]

  cleaned_text = ' '.join(words)
  return cleaned_text

  blog_df['Cleaned_Data'] = blog_df['Data'].apply(clean_text)


  tfidf_vectorizer = TfidfVectorizer(max_features=5000)
  X_tfidf = tfidf_vectorizer.fit_transform(blog_df['Cleaned_Data'])


  cleaned_data_head = blog_df[['Data', 'Cleaned_Data']].head()
  tfidf_shape = X_tfidf.shape

  cleaned_data_head, tfidf_shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix


```

```python
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(blog_df['Labels'])


# Splitting the data into training and test sets

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, blog_df['Labels'], test_size=0.2, random_state=42)

# Implementing the Naive Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)

# Making predictions on the test set
y_pred = nb_classifier.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
confusion_mat = confusion_matrix(y_test, y_pred)

accuracy, classification_rep, confusion_mat
```
